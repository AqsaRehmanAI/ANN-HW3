### (a) Repeat HW2 with RNN, you are free to design your own network architecture. But you need to have at least one LSTM layer.

```python
# ===============================
# HW3 - Part (a): LSTM Model 
# ===============================

import pandas as pd
import numpy as np
import yfinance as yf
import torch
import torch.nn as nn
import torch.nn.functional as F
import pytorch_lightning as pl
from torch.utils.data import Dataset, DataLoader, random_split
from pytorch_lightning.callbacks import EarlyStopping, ModelCheckpoint
from pytorch_lightning.loggers import CSVLogger
import matplotlib.pyplot as plt
import os
from pathlib import Path

# -----------------------------------------
# 1. Dataset Preparation 
# -----------------------------------------

def download_stock_data(tickers, start_date, end_date=None):

    data = {}
    for ticker in tickers:
        try:
            df = yf.Ticker(ticker).history(start=start_date, end=end_date)
            if len(df) == 0:
                raise ValueError(f"No data for {ticker}")
            data[ticker] = df['Close'].values
        except Exception as e:
            print(f"Error downloading {ticker}: {str(e)}")
            # Create dummy data if download fails
            dummy_length = len(data.get(tickers[0], [])) if data else 100
            data[ticker] = np.random.rand(dummy_length).astype(np.float32)
    
    # Ensure all arrays have same length
    max_len = max(len(v) for v in data.values())
    for ticker in tickers:
        if len(data[ticker]) < max_len:
            data[ticker] = np.pad(data[ticker], (0, max_len - len(data[ticker])))
    
    return pd.DataFrame(data)


feature_stocks = ['tsla','meta','nvda','amzn','nflx','gbtc','gdx','intc',
                 'dal','c','goog','aapl','msft','ibm','hp','orcl','sap','crm','hubs','twlo']
predict_stock = 'msft'
start_date = '2020-01-01'

print("Downloading data...")
X_raw = download_stock_data(feature_stocks, start_date)
y_raw = download_stock_data([predict_stock], start_date)


X_data = X_raw.values.T.astype(np.float32)  # (features, time)
y_data = y_raw.values.astype(np.float32)    # (time, 1)

# -----------------------------------------
# 2. Define Dataset Class
# -----------------------------------------

class StockDataset(Dataset):
    def __init__(self, X, Y, days=5):
        self.X = X
        self.Y = Y.reshape(-1)
        self.days = days

    def __len__(self):
        return len(self.Y) - self.days

    def __getitem__(self, idx):
        x = self.X[:, idx:idx+self.days]  # Shape: (features, days)
        y = self.Y[idx+self.days]         # Target
        return torch.FloatTensor(x), torch.FloatTensor([y])

# Parameters
days = 5  # Use last 5 days
dataset = StockDataset(X_data, y_data, days=days)

# -----------------------------------------
# 3. DataModule
# -----------------------------------------

train_size = int(0.7 * len(dataset))
val_size = int(0.15 * len(dataset))
test_size = len(dataset) - train_size - val_size

train_set, val_set, test_set = random_split(dataset, [train_size, val_size, test_size],
                                          generator=torch.Generator().manual_seed(42))

class StockDataModule(pl.LightningDataModule):
    def __init__(self, train, val, test, batch_size=32):
        super().__init__()
        self.train_set = train
        self.val_set = val
        self.test_set = test
        self.batch_size = batch_size

    def train_dataloader(self):
        return DataLoader(self.train_set, batch_size=self.batch_size, shuffle=True)

    def val_dataloader(self):
        return DataLoader(self.val_set, batch_size=self.batch_size)

    def test_dataloader(self):
        return DataLoader(self.test_set, batch_size=self.batch_size)

data_module = StockDataModule(train_set, val_set, test_set)

# -----------------------------------------
# 4. LSTM Model (Improved Version)
# -----------------------------------------

class LSTMPricePredictor(pl.LightningModule):
    def __init__(self, input_size=20, hidden_size=64, lstm_layers=1, dropout_rate=0.3, learning_rate=1e-3):
        super().__init__()
        self.save_hyperparameters()
        
        # LSTM layers
        self.lstm = nn.LSTM(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=lstm_layers,
            batch_first=True,
            dropout=dropout_rate if lstm_layers > 1 else 0
        )
        self.dropout = nn.Dropout(dropout_rate)
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        # x shape:
        x = x.permute(0, 2, 1)  
        
        # LSTM
        lstm_out, _ = self.lstm(x)
        last_out = lstm_out[:, -1, :]  # Take last output
        last_out = self.dropout(last_out)
        
        return self.fc(last_out)

    def training_step(self, batch, batch_idx):
        x, y = batch
        y_hat = self(x)
        loss = F.mse_loss(y_hat, y)
        self.log('train_loss', loss, prog_bar=True)
        return loss

    def validation_step(self, batch, batch_idx):
        x, y = batch
        y_hat = self(x)
        loss = F.mse_loss(y_hat, y)
        self.log('val_loss', loss, prog_bar=True)
        return loss

    def test_step(self, batch, batch_idx):
        x, y = batch
        y_hat = self(x)
        loss = F.mse_loss(y_hat, y)
        self.log('test_loss', loss)
        return loss

    def configure_optimizers(self):
        optimizer = torch.optim.Adam(self.parameters(), lr=self.hparams.learning_rate)
        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=5)
        return {
            'optimizer': optimizer,
            'lr_scheduler': {
                'scheduler': scheduler,
                'monitor': 'val_loss'
            }
        }

# Instantiate model
model = LSTMPricePredictor(
    input_size=20,
    hidden_size=128,  # Increased capacity
    lstm_layers=2,    # Stacked LSTM
    dropout_rate=0.2,
    learning_rate=1e-3
)

# -----------------------------------------
# 5. Training Setup 
# -----------------------------------------

log_dir = os.path.abspath("lstm_logs")
os.makedirs(log_dir, exist_ok=True)

# Callbacks
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=15,
    mode='min',
    verbose=True
)

checkpoint = ModelCheckpoint(
    dirpath=log_dir,
    filename='best_model',
    monitor='val_loss',
    mode='min',
    save_top_k=1
)

logger = CSVLogger(save_dir=log_dir, name='', version='')

# Trainer
trainer = pl.Trainer(
    max_epochs=100,
    callbacks=[early_stop, checkpoint],
    logger=logger,
    deterministic=True,
    enable_progress_bar=True,
    log_every_n_steps=5,
    accelerator='auto'
)

# Train
print("\nStarting training...")
trainer.fit(model, datamodule=data_module)

# Test
print("\nEvaluating on test set...")
test_result = trainer.test(datamodule=data_module, ckpt_path='best')
print(f"Test MSE: {test_result[0]['test_loss']:.4f}")

# -----------------------------------------
# 6. Results and Plotting 
# -----------------------------------------

# Get the correct metrics path
metrics_path = os.path.join(logger.log_dir, "metrics.csv")

if os.path.exists(metrics_path):
    metrics = pd.read_csv(metrics_path)
    
    # Clean and plot
    train_loss = metrics['train_loss'].dropna().reset_index(drop=True)
    val_loss = metrics['val_loss'].dropna().reset_index(drop=True)

    plt.figure(figsize=(10, 5))
    plt.plot(train_loss, label='Train Loss')
    plt.plot(val_loss, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('MSE Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.grid()
    plt.show()

    # Print final metrics
    final_metrics = {
        'Final Train Loss': train_loss.iloc[-1],
        'Final Val Loss': val_loss.iloc[-1],
        'Test Loss': test_result[0]['test_loss']
    }

    print("\nFinal Metrics:")
    for k, v in final_metrics.items():
        print(f"{k}: {v:.4f}")
else:
    print("\nWarning: Could not find metrics.csv. Plotting skipped.")

print("\nTraining complete!")
```

```python
# ===============================
# HW3 - Part (c): Knowledge Distillation 
# ===============================

import pandas as pd
import numpy as np
import yfinance as yf
import torch
import torch.nn as nn
import torch.nn.functional as F
import pytorch_lightning as pl
from torch.utils.data import Dataset, DataLoader, random_split
from pytorch_lightning.callbacks import EarlyStopping, ModelCheckpoint
from pytorch_lightning.loggers import CSVLogger
import matplotlib.pyplot as plt
import os

# -----------------------------------------
# 1. Dataset Preparation
# -----------------------------------------

def download_stock_data(tickers, start_date, end_date=None):
    """Improved data download function with error handling"""
    data = {}
    for ticker in tickers:
        try:
            df = yf.Ticker(ticker).history(start=start_date, end=end_date)
            if len(df) == 0:
                raise ValueError(f"No data for {ticker}")
            data[ticker] = df['Close'].values
        except Exception as e:
            print(f"Error downloading {ticker}: {str(e)}")
            dummy_length = len(data.get(tickers[0], [])) if data else 100
            data[ticker] = np.random.rand(dummy_length).astype(np.float32)
    
    max_len = max(len(v) for v in data.values())
    for ticker in tickers:
        if len(data[ticker]) < max_len:
            data[ticker] = np.pad(data[ticker], (0, max_len - len(data[ticker])))
    
    return pd.DataFrame(data)

feature_stocks = ['tsla','meta','nvda','amzn','nflx','gbtc','gdx','intc',
                 'dal','c','goog','aapl','msft','ibm','hp','orcl','sap','crm','hubs','twlo']
predict_stock = 'msft'
start_date = '2020-01-01'

print("Downloading data...")
X_raw = download_stock_data(feature_stocks, start_date)
y_raw = download_stock_data([predict_stock], start_date)

X_data = X_raw.values.T.astype(np.float32)
y_data = y_raw.values.astype(np.float32)

# -----------------------------------------
# 2. Dataset Class
# -----------------------------------------

class StockDataset(Dataset):
    def __init__(self, X, Y, days=5):
        self.X = X
        self.Y = Y.reshape(-1)
        self.days = days

    def __len__(self):
        return len(self.Y) - self.days

    def __getitem__(self, idx):
        x = self.X[:, idx:idx+self.days]
        y = self.Y[idx+self.days]
        return torch.FloatTensor(x), torch.FloatTensor([y])

days = 5
dataset = StockDataset(X_data, y_data, days=days)

# -----------------------------------------
# 3. DataModule
# -----------------------------------------

train_size = int(0.7 * len(dataset))
val_size = int(0.15 * len(dataset))
test_size = len(dataset) - train_size - val_size

train_set, val_set, test_set = random_split(dataset, [train_size, val_size, test_size],
                                            generator=torch.Generator().manual_seed(42))

class StockDataModule(pl.LightningDataModule):
    def __init__(self, train, val, test, batch_size=32):
        super().__init__()
        self.train_set = train
        self.val_set = val
        self.test_set = test
        self.batch_size = batch_size

    def train_dataloader(self):
        return DataLoader(self.train_set, batch_size=self.batch_size, shuffle=True)

    def val_dataloader(self):
        return DataLoader(self.val_set, batch_size=self.batch_size)

    def test_dataloader(self):
        return DataLoader(self.test_set, batch_size=self.batch_size)

data_module = StockDataModule(train_set, val_set, test_set)

# -----------------------------------------
# 4. Define Teacher and Student Models
# -----------------------------------------

class LSTMPricePredictor(pl.LightningModule):
    def __init__(self, input_size=20, hidden_size=128, lstm_layers=2, dropout_rate=0.2, learning_rate=1e-3):
        super().__init__()
        self.save_hyperparameters()
        self.lstm = nn.LSTM(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=lstm_layers,
            batch_first=True,
            dropout=dropout_rate if lstm_layers > 1 else 0
        )
        self.dropout = nn.Dropout(dropout_rate)
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        x = x.permute(0, 2, 1)  # (batch_size, days, features)
        out, _ = self.lstm(x)
        out = out[:, -1, :]
        out = self.dropout(out)
        return self.fc(out)

    def training_step(self, batch, batch_idx):
        x, y = batch
        y_hat = self(x)
        loss = F.mse_loss(y_hat, y)
        self.log('train_loss', loss)
        return loss

    def validation_step(self, batch, batch_idx):
        x, y = batch
        y_hat = self(x)
        loss = F.mse_loss(y_hat, y)
        self.log('val_loss', loss)
        return loss

    def test_step(self, batch, batch_idx):
        x, y = batch
        y_hat = self(x)
        loss = F.mse_loss(y_hat, y)
        self.log('test_loss', loss)
        return loss

    def configure_optimizers(self):
        optimizer = torch.optim.Adam(self.parameters(), lr=self.hparams.learning_rate)
        return optimizer

# Student Model (smaller model)
class StudentLSTM(pl.LightningModule):
    def __init__(self, teacher, input_size=20, hidden_size=64, lstm_layers=1, dropout_rate=0.2, learning_rate=1e-3, alpha=0.5):
        super().__init__()
        self.save_hyperparameters()
        self.teacher = teacher.eval()  # Freeze teacher
        for param in self.teacher.parameters():
            param.requires_grad = False

        self.lstm = nn.LSTM(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=lstm_layers,
            batch_first=True
        )
        self.dropout = nn.Dropout(dropout_rate)
        self.fc = nn.Linear(hidden_size, 1)

    def forward(self, x):
        x = x.permute(0, 2, 1)
        out, _ = self.lstm(x)
        out = out[:, -1, :]
        out = self.dropout(out)
        return self.fc(out)

    def training_step(self, batch, batch_idx):
        x, y = batch
        with torch.no_grad():
            teacher_preds = self.teacher(x)

        student_preds = self(x)

        loss_true = F.mse_loss(student_preds, y)
        loss_teacher = F.mse_loss(student_preds, teacher_preds)

        loss = self.hparams.alpha * loss_true + (1 - self.hparams.alpha) * loss_teacher
        self.log('train_loss', loss)
        return loss

    def validation_step(self, batch, batch_idx):
        x, y = batch
        student_preds = self(x)
        loss = F.mse_loss(student_preds, y)
        self.log('val_loss', loss)
        return loss

    def test_step(self, batch, batch_idx):
        x, y = batch
        student_preds = self(x)
        loss = F.mse_loss(student_preds, y)
        self.log('test_loss', loss)
        return loss

    def configure_optimizers(self):
        optimizer = torch.optim.Adam(self.parameters(), lr=self.hparams.learning_rate)
        return optimizer

# -----------------------------------------
# 5. Load Trained Teacher 
# -----------------------------------------

# Initialize and Train Teacher if needed
teacher = LSTMPricePredictor()
trainer_teacher = pl.Trainer(max_epochs=20, accelerator='auto', enable_progress_bar=True)
print("\nTraining Teacher Model...")
trainer_teacher.fit(teacher, datamodule=data_module)

# Evaluate teacher
print("\nEvaluating Teacher Model...")
trainer_teacher.test(teacher, datamodule=data_module)

# -----------------------------------------
# 6. Train Student with Knowledge Distillation
# -----------------------------------------


student = StudentLSTM(teacher)


log_dir = os.path.abspath("distillation_logs")
os.makedirs(log_dir, exist_ok=True)
logger = CSVLogger(save_dir=log_dir, name='', version='')


trainer_student = pl.Trainer(
    max_epochs=100,
    callbacks=[
        EarlyStopping(monitor='val_loss', patience=15, mode='min'),
        ModelCheckpoint(monitor='val_loss', save_top_k=1, mode='min')
    ],
    logger=logger,
    deterministic=True,
    accelerator='auto'
)

# Train Student
print("\nTraining Student Model with Knowledge Distillation...")
trainer_student.fit(student, datamodule=data_module)

# Test Student
print("\nEvaluating Student Model...")
test_result = trainer_student.test(student, datamodule=data_module, ckpt_path='best')

# -----------------------------------------
# 7. Plot Results
# -----------------------------------------

metrics_path = os.path.join(logger.log_dir, "metrics.csv")

if os.path.exists(metrics_path):
    metrics = pd.read_csv(metrics_path)
    
    train_loss = metrics['train_loss'].dropna().reset_index(drop=True)
    val_loss = metrics['val_loss'].dropna().reset_index(drop=True)

    plt.figure(figsize=(10, 5))
    plt.plot(train_loss, label='Train Loss')
    plt.plot(val_loss, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('MSE Loss')
    plt.title('Student Model - Training and Validation Loss (Knowledge Distillation)')
    plt.legend()
    plt.grid()
    plt.show()

    print("\nFinal Metrics:")
    print(f"Final Train Loss: {train_loss.iloc[-1]:.4f}")
    print(f"Final Val Loss: {val_loss.iloc[-1]:.4f}")
    print(f"Test Loss: {test_result[0]['test_loss']:.4f}")

else:
    print("\nWarning: metrics.csv not found. Skipping plot.")

print("\nTraining complete!")
```

```python
# ===============================
# HW3 - Part (d): Mamba Network 
# ===============================

import pandas as pd
import numpy as np
import yfinance as yf
import torch
import torch.nn as nn
import torch.nn.functional as F
import pytorch_lightning as pl
from torch.utils.data import Dataset, DataLoader, random_split
from pytorch_lightning.callbacks import EarlyStopping, ModelCheckpoint
from pytorch_lightning.loggers import CSVLogger
import matplotlib.pyplot as plt
import os

# -----------------------------------------
# 1. Dataset Preparation 
# -----------------------------------------

def download_stock_data(tickers, start_date, end_date=None):
    data = {}
    for ticker in tickers:
        try:
            df = yf.Ticker(ticker).history(start=start_date, end=end_date)
            if len(df) == 0:
                raise ValueError(f"No data for {ticker}")
            data[ticker] = df['Close'].values
        except Exception as e:
            print(f"Error downloading {ticker}: {str(e)}")
            dummy_length = len(data.get(tickers[0], [])) if data else 100
            data[ticker] = np.random.rand(dummy_length).astype(np.float32)

    max_len = max(len(v) for v in data.values())
    for ticker in tickers:
        if len(data[ticker]) < max_len:
            data[ticker] = np.pad(data[ticker], (0, max_len - len(data[ticker])))

    return pd.DataFrame(data)

feature_stocks = ['tsla','meta','nvda','amzn','nflx','gbtc','gdx','intc',
                 'dal','c','goog','aapl','msft','ibm','hp','orcl','sap','crm','hubs','twlo']
predict_stock = 'msft'
start_date = '2020-01-01'

print("Downloading data...")
X_raw = download_stock_data(feature_stocks, start_date)
y_raw = download_stock_data([predict_stock], start_date)

X_data = X_raw.values.T.astype(np.float32)
y_data = y_raw.values.astype(np.float32)

# -----------------------------------------
# 2. Dataset Class
# -----------------------------------------

class StockDataset(Dataset):
    def __init__(self, X, Y, days=5):
        self.X = X
        self.Y = Y.reshape(-1)
        self.days = days

    def __len__(self):
        return len(self.Y) - self.days

    def __getitem__(self, idx):
        x = self.X[:, idx:idx+self.days]
        y = self.Y[idx+self.days]
        return torch.FloatTensor(x), torch.FloatTensor([y])

days = 5
dataset = StockDataset(X_data, y_data, days=days)

# -----------------------------------------
# 3. DataModule
# -----------------------------------------

train_size = int(0.7 * len(dataset))
val_size = int(0.15 * len(dataset))
test_size = len(dataset) - train_size - val_size

train_set, val_set, test_set = random_split(dataset, [train_size, val_size, test_size],
                                            generator=torch.Generator().manual_seed(42))

class StockDataModule(pl.LightningDataModule):
    def __init__(self, train, val, test, batch_size=32):
        super().__init__()
        self.train_set = train
        self.val_set = val
        self.test_set = test
        self.batch_size = batch_size

    def train_dataloader(self):
        return DataLoader(self.train_set, batch_size=self.batch_size, shuffle=True)

    def val_dataloader(self):
        return DataLoader(self.val_set, batch_size=self.batch_size)

    def test_dataloader(self):
        return DataLoader(self.test_set, batch_size=self.batch_size)

data_module = StockDataModule(train_set, val_set, test_set)

# -----------------------------------------
# 4. Simple Mamba Layer
# -----------------------------------------

class SimpleMambaLayer(nn.Module):
    def __init__(self, embed_dim):
        super().__init__()
        self.conv1 = nn.Conv1d(embed_dim, embed_dim, kernel_size=3, padding=1)
        self.conv2 = nn.Conv1d(embed_dim, embed_dim, kernel_size=3, padding=1)
        self.activation = nn.GELU()
        self.dropout = nn.Dropout(0.2)

    def forward(self, x):
        # x: (batch_size, seq_len, embed_dim)
        x = x.permute(0, 2, 1)  # (batch, embed_dim, seq_len)
        x = self.conv1(x)
        x = self.activation(x)
        x = self.conv2(x)
        x = self.activation(x)
        x = self.dropout(x)
        x = x.permute(0, 2, 1)  # (batch, seq_len, embed_dim)
        return x

# -----------------------------------------
# 5. Mamba Network Model
# -----------------------------------------

class MambaPricePredictor(pl.LightningModule):
    def __init__(self, input_size=20, embed_dim=64, learning_rate=1e-3):
        super().__init__()
        self.save_hyperparameters()

        self.input_proj = nn.Linear(input_size, embed_dim)
        self.mamba_layer = SimpleMambaLayer(embed_dim)
        self.fc = nn.Sequential(
            nn.Linear(embed_dim, 32),
            nn.ReLU(),
            nn.Linear(32, 1)
        )

    def forward(self, x):
        # x: (batch_size, features, days)
        x = x.permute(0, 2, 1)  # (batch_size, days, features)
        x = self.input_proj(x)  # (batch_size, days, embed_dim)
        x = self.mamba_layer(x)  # (batch_size, days, embed_dim)
        x = x[:, -1, :]  # Use last time step
        return self.fc(x)

    def training_step(self, batch, batch_idx):
        x, y = batch
        y_hat = self(x)
        loss = F.mse_loss(y_hat, y)
        self.log('train_loss', loss)
        return loss

    def validation_step(self, batch, batch_idx):
        x, y = batch
        y_hat = self(x)
        loss = F.mse_loss(y_hat, y)
        self.log('val_loss', loss)
        return loss

    def test_step(self, batch, batch_idx):
        x, y = batch
        y_hat = self(x)
        loss = F.mse_loss(y_hat, y)
        self.log('test_loss', loss)
        return loss

    def configure_optimizers(self):
        optimizer = torch.optim.Adam(self.parameters(), lr=self.hparams.learning_rate)
        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=5)
        return {
            'optimizer': optimizer,
            'lr_scheduler': {
                'scheduler': scheduler,
                'monitor': 'val_loss'
            }
        }

model = MambaPricePredictor(input_size=20, embed_dim=64)

# -----------------------------------------
# 6. Training Setup
# -----------------------------------------

log_dir = os.path.abspath("mamba_logs")
os.makedirs(log_dir, exist_ok=True)

early_stop = EarlyStopping(monitor='val_loss', patience=15, mode='min', verbose=True)
checkpoint = ModelCheckpoint(dirpath=log_dir, filename='best_model', monitor='val_loss', mode='min', save_top_k=1)
logger = CSVLogger(save_dir=log_dir, name='', version='')

trainer = pl.Trainer(
    max_epochs=100,
    callbacks=[early_stop, checkpoint],
    logger=logger,
    deterministic=True,
    enable_progress_bar=True,
    log_every_n_steps=5,
    accelerator='auto'
)

# Train
print("\nStarting training...")
trainer.fit(model, datamodule=data_module)

# Test
print("\nEvaluating on test set...")
test_result = trainer.test(model, datamodule=data_module, ckpt_path='best')

# -----------------------------------------
# 7. Plot Results
# -----------------------------------------

metrics_path = os.path.join(logger.log_dir, "metrics.csv")

if os.path.exists(metrics_path):
    metrics = pd.read_csv(metrics_path)

    train_loss = metrics['train_loss'].dropna().reset_index(drop=True)
    val_loss = metrics['val_loss'].dropna().reset_index(drop=True)

    plt.figure(figsize=(10, 5))
    plt.plot(train_loss, label='Train Loss')
    plt.plot(val_loss, label='Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('MSE Loss')
    plt.title('Training and Validation Loss (Mamba Network)')
    plt.legend()
    plt.grid()
    plt.show()

    print("\nFinal Metrics:")
    print(f"Final Train Loss: {train_loss.iloc[-1]:.4f}")
    print(f"Final Val Loss: {val_loss.iloc[-1]:.4f}")
    print(f"Test Loss: {test_result[0]['test_loss']:.4f}")

else:
    print("\nWarning: metrics.csv not found. Skipping plot.")

print("\nTraining complete!")
```
